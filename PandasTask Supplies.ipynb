11. Gorup by region and find out which region is more profitable

```python
OrdersDf[['Region','Profit']].groupby(['Region']).aggregate({'Profit':'sum'}).reset_index().sort_values('Profit', ascending=False).head(1)
```

12. Try to find out overalll in which country we are giving more didscount

```python
OrdersDf[['State or Province','Discount']].groupby(['State or Province']).aggregate({'Discount':'sum'}).reset_index().sort_values('Discount',ascending=False)
```
